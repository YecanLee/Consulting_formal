```python
import webdataset as wds
import json
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

def show_sample_from_shard(shard_path, max_images: int = 10):
    dataset = wds.WebDataset(shard_path).decode("pil").to_tuple("jpg", "json")

    for idx, (image, json_bytes) in enumerate(dataset):
        description = json_bytes["description"]
        display(Markdown(f"### ðŸŽ¨ Image {idx + 1} Description\n{description}"))
        plt.imshow(image)
        plt.axis("off")
        plt.show()
    
        if idx + 1 >= max_images:
            break

```

```python
show_sample_from_shard("web_dataset/baroque-00000.tar")
```
